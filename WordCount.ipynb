```python
from pyspark import SparkContext
```

```python
sc = SparkContext(master = "local[*]", appName = "wordcount")
```

```python
log_rdd = sc.textFile("/content/Impute data")
```

```python
log_rdd.take(10)
```

```python
log_rdd.count()
```

```python
filtered_log_rdd = log_rdd.filter(lambda x: "tower" in x)
```

```python
filtered_log_rdd.collect()
```

```python
Word_rdd=log_rdd.flatMap(lambda x: x.split())
```

```python
Word_rdd.take(10)
```

```python
filtered_word_rdd = Word_rdd.filter(lambda x: "tower" in x)
```

```python
filtered_word_rdd.collect()
```

```python
word_pairs_rdd = Word_rdd.map(lambda word: (word.lower(), 1))
```

```python
word_counts_rdd = word_pairs_rdd.reduceByKey(lambda a, b: a + b)
```

```python
word_counts = word_counts_rdd.collect()
```
